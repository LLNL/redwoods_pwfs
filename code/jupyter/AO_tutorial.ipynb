**Section 1 Imaging through the turbulence**

```python
# diffraction limiting resolution of human eye
import numpy as np
import matplotlib.pyplot as plt

def resolution(lam, D):
    # takes in wavelength of interest and diameter of aperature
    # returns resolution in m
    return 1.22*lam/D

pupil_d = 4 *10**(-3) # diameter in mm
lam = np.linspace(380, 780, (780-380))*10**(-9) # look at typical range of human sight

res = resolution(lam, pupil_d)

plt.figure()
plt.plot(lam, res)
plt.ylabel('resolution')
plt.xlabel('wavelength $\lambda$')
plt.tight_layout()
plt.show()


```
