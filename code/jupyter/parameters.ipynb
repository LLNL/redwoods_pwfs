```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/transmit.csv')
wavelength = np.array(df['wavelength'])
percent = np.array(df[' percent'])

index = np.where((wavelength>= 950) & (wavelength<=1600))

plt.figure()
plt.plot(wavelength[index], percent[index])
plt.show()

print(f'Average atm percent: {np.mean(percent[index])}')
```
