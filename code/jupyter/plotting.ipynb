```python
import matplotlib.pyplot as plt

# Sample data points
x_points = [1, 2, 3, 4, 5]
y_points = [2, 4, 1, 3, 5]

# Line properties
line_width = 2
line_color = 'red'

# Create the plot
fig, ax = plt.subplots()
ax.scatter(x_points, y_points, color='blue')  # Plot the points

# Add horizontal lines and labels
for x, y in zip(x_points, y_points):
    ax.hlines(y, x - 0.2, x + 0.2, linewidth=line_width, color=line_color)
    ax.text(x, y + 0.2, f'{y}', ha='center', va='bottom')  # Adjust y + 0.2 for label position

# Set plot labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Horizontal Lines with Labels')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# create a grid
x = np.linspace(0, 2*np.pi, 128)
y = np.linspace(0, 2*np.pi, 128)
X, Y = np.meshgrid(x, y)

# define high freq diagonal mode
modes = np.sin(5*X - 2*Y) 

# Plot
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
im = ax.imshow(modes)
plt.tight_layout()
plt.show()
```
